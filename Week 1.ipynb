```python
#Exercise 1.1 a)
sigma1= 1/2
sigma2= 1/2
mu1 = 0
mu2 = 1
d1 = np.random.normal(loc=0,scale=1/2, size=10000)
d2 = np.random.normal(loc=1,scale=sigma2, size=10000)
count1, bins1, ignored1 = plt.hist(d1,30,density=True)
count2, bins2, ignored2 = plt.hist(d2,30,density=True)
plt.plot(bins1, 1/(sigma1 *np.sqrt(2*np.pi)) * np.exp(- (bins1-mu1)**2/(2*sigma1**2)), linewidth=2, color='r')
plt.plot(bins2, 1/(sigma2 *np.sqrt(2*np.pi)) * np.exp(- (bins2-mu2)**2/(2*sigma2**2)), linewidth=2, color='r')

print("looks like approximately at 1/2 they intersect")
```
