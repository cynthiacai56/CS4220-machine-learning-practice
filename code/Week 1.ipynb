## 1.6 Density estimation using Parzen densities
Next to classifiers, PRTools also has the possibility to estimate densities. In this section we are going to estimate the density using a Parzen density estimator, called parzenm in PRTools.

### 1.6.1 The density estimation with different h
h: the width parameter

For the given data (Simple dataset), plot the Parzen density estimate for different values of h. What is the best value of h?

```python
# Exercise 1.17
a = pr.gendats([20,20],1,8)
h_list =[0.05,0.1,0.3,0.5,1,2,3]

a = pr.prdataset(+a)
for h in h_list:
    w = pr.parzenm(a,h)

    plt.figure()
    pr.scatterd(a)
    #pr.plotm(w)
    pr.plotm(w,gridsize=100)
```

### 1.6.2 Log-likelihood
For Simple dataset, what is the optimal value for h, i.e. the maximal likelihood? Is this also the best density estimate for the dataset?

Also, if you want to evaluate the performance of an estimated ^p, use an independent test set!

```python
# Exercise 1.18
a = pr.gendats([20,20],1,8) # Generate data
a = pr.prdataset(+a)
hs = [0.01,0.05,0.1,0.25,0.5,1,1.5,2,3,4,5] # Array of h's to try
LL = np.zeros(len(hs)) # LL means log-likelihood
for i in range(len(hs)): # For each h...
    w = pr.parzenm(a,hs[i]) # estimate Parzen density
    LL[i] = np.sum(np.log(+(a*w))) # calculate log-likelihood
plt.plot(hs,LL) # plot log-likelihood as function of h
```

```python
# Exercise 1.19
[trn,tst] = pr.gendat(a,0.5) # Split into trn and tst, both 50%
hs = [0.01,0.05,0.1,0.25,0.5,1,1.5,2,3,4,5] # Array of h's to try
Ltrn = np.zeros(len(hs))
Ltst = np.zeros(len(hs))
for i in range(len(hs)): # For each h...
    w = pr.parzenm(trn,hs[i]) # estimate Parzen density
    Ltrn[i] = np.sum(np.log(+(trn*w))) # calculate trn log-likelihood
    Ltst[i] = np.sum(np.log(+(tst*w))) # calculate tst log-likelihood

plt.plot(hs,Ltrn,'b-') # Plot trn log-likelihood as function of h
plt.plot(hs,Ltst,'r-') # Plot tst log-likelihood as function
```

```python
# Linear discriminant analysis
ldc = pr.ldc()
ldc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(ldc)

ldc = pr.ldc()
ldc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(ldc)
```

```python
# Quadratic discriminant analysis
qdc = pr.qdc()
qdc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(qdc)

qdc = pr.qdc()
qdc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(qdc)
```

```python
# Nearest mean classifier
nmc = pr.nmc()
nmc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(nmc)

nmc = pr.nmc()
nmc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(nmc)
```

```python
# Fishers linear discriminant
fisherc = pr.fisherc()
fisherc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(fisherc)

fisherc = pr.fisherc()
fisherc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(fisherc)
```

```python
# K-nearest neighbour classifier
knnc = pr.knnc() # k=1 ???
knnc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(knnc)

knnc = pr.knnc()
knnc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(knnc)
```

```python
# Parzen classifier
parzenc = pr.parzenc()
parzenc.train(banana)
plt.figure()
pr.scatterd(banana)
pr.plotc(parzenc)

parzenc = pr.parzenc()
parzenc.train(scale_banana)
plt.figure()
pr.scatterd(scale_banana)
pr.plotc(parzenc)
```
